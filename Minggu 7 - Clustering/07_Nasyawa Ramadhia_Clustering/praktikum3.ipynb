**Import Library**

```python
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

menggunakan dataset Iris

```python
# Load dataset
iris = datasets.load_iris()
data = iris.data
```

normalisasi agar nilai rentang seragam

```python
# Normalisasi data
data = data / data.max(axis=0)
```

menentukan ukuran peta, (10*10), inisialisasi SOM

```python
# Inisialisasi SOM
map_size = (10, 10)
som = MiniSom(map_size[0], map_size[1], data.shape[1], sigma=0.5, learning_rate=0.5)
```

Inisialisasi bobot SOM secara acak.

metode yang digunakan untuk menginisialisasi bobot (weights) dari unit-unit(neuron) dalam peta Kohonen dengan nilai acak.

```python
som.random_weights_init(data)
```

Melatih SOM dengan data menggunakan metode train_random

bobot-bobot unit SOM akan diperbarui berdasarkan data input secara acak atau stokastik untuk mencapai konvergensi. Proses ini akan menghasilkan peta Kohonen yang dapat digunakan untuk pemetaan atau klasifikasi data lebih lanjut.

```python
# Pelatihan SOM
num_epochs = 100   # jumlah iterasi atau epoch
som.train_random(data, num_epochs) 
#melatih SOM menggunakan data yang diberikan selama sejumlah epoch yang ditentukan
```

Menggunakan peta hasil pelatihan untuk visualisasi

```python
# Visualisasi hasil SOM
plt.figure(figsize=(8, 8)) #gambar uk 8x8 
for i, x in enumerate(data):
    w = som.winner(x)  # Pemenang untuk sampel x
    plt.text(w[0]+.5, w[1]+.5, str(i+1), color='k', fontdict={'weight': 'bold', 'size': 11}) #mengatur tampilan teks.
plt.pcolor(som.distance_map().T, cmap='bone_r', alpha=.2) #membuat peta warna
plt.colorbar() # menambahkan skala warna

plt.show() #menampilkan
```
